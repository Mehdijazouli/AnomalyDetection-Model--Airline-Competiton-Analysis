```python
import pandas as pd

df = pd.read_excel('route_circuity.xlsx')
df
```

```python
ville_mar = ["AGA" ,"CMN" ,"VIL" ,"ESU", "FEZ" ,"RAK" ,"NDR" ,"OZZ" ,"OUD" ,"RBA" ,"TNG"]
```

```python
# Supprimer les lignes ne rÃ©pondant pas aux critÃ¨res
condition = ((df['Circuity'] >= 110) & (df['Circuity'] <= 130)) | ((df['Arr'].isin(ville_mar)) | (df['Dep'].isin(ville_mar)))
df = df[condition]
df = df[df['Mkt Al'] != 'AT']
```

```python
df["Agence"] = df["Mkt Al"]
df["Route_base"] = df["Route"]
df
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder

df["Agence"] = df["Mkt Al"]
df["Route_base"] = df["Route"]

label_encoder = LabelEncoder()
df["Mkt Al"] = label_encoder.fit_transform(df["Mkt Al"])
df["Route"] = label_encoder.fit_transform(df["Route"])

# SÃ©lectionner les colonnes pertinentes
columns = ["Somme de Ops", "Mkt Al", "Route"]

# CrÃ©er et entraÃ®ner le modÃ¨le Isolation Forest Extended
model = IsolationForest(contamination=0.05, random_state=42, bootstrap=True)
model.fit(df[columns])

# PrÃ©dire les anomalies
df["Anomalies"] = model.predict(df[columns])

# Les valeurs de "Somme de Ops" considÃ©rÃ©es comme anomalies sont celles oÃ¹ "Anomalies" est Ã©gal Ã  -1
anomalies = df[df["Anomalies"] == -1]
```
